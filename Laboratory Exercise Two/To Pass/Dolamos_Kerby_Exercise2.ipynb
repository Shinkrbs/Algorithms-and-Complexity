# Sorting Algorithm Running Times vs. Input Size

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(10000, 100001, 10000)

# Read data from files
selection = np.loadtxt('SelectionSortTimes.txt')  
bubble = np.loadtxt('BubbleSortTimes.txt')       
insertion = np.loadtxt('InsertionSortTimes.txt') 
merge = np.loadtxt('MergeSortTimes.txt')         
quick = np.loadtxt('QuickSortTimes.txt')

plt.figure(figsize=(10, 7))

plt.plot(x, bubble, marker='o', c='r', label='Bubble Sort O(n)')
plt.plot(x, insertion, marker='o', c='b', label='Insertion Sort O(n)')
plt.plot(x, selection, marker='o', c='orange', label='Selection Sort O(n)')
plt.plot(x, merge, marker='o', c='g', label='Merge Sort O(nlogn)')
plt.plot(x, quick, marker='o', c='y', label='Quick Sort O(nlogn)')

plt.title("Sorting Algorithm Running Times vs Input Size")
plt.xlabel("Input Size (n)")
plt.ylabel("Running Time (Seconds)")

plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```

The graph shows the running times of five sorting algorithms: Bubble Sort, Insertion Sort, Selection Sort, Merge Sort, and Quick Sort. The experiment used five sets of vectors, each containing 20 vectors with sizes ranging from 10,000 to 100,000, increasing by 10,000 in each step. The data used for sorting was completely random, making the test more realistic and accurately showing how each algorithm performs.  

From the graph, we can see that sorting algorithms with a time complexity of O(n) are much slower than those with O(n log n). This proves that algorithms using the Divide and Conquer approach, like Merge Sort and Quick Sort, are much faster and more efficient.  

Among the O(n) algorithms, Insertion Sort is the fastest, while Bubble Sort is the slowest not just among the O(n) algorithms but out of all five sorting methods tested.

# O(nlogn) Sorting Algorithm Graph

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(10000, 100001, 10000)

# Read data from files
merge = np.loadtxt('MergeSortTimes.txt')         
quick = np.loadtxt('QuickSortTimes.txt')

# Plot Merge Sort vs Quick Sort
plt.figure(figsize=(10, 7))
plt.plot(x, merge, marker='s', c='orange', label='Merge Sort O(n log n)')
plt.plot(x, quick, marker='d', c='maroon', label='Quick Sort O(n log n)')

plt.title("Merge Sort vs Quick Sort Running Times")
plt.xlabel("Input Size (n)")
plt.ylabel("Running Time (Seconds)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

The comparison between Quick Sort and Merge Sort shows that Quick Sort is generally faster. For **10,000 elements**, Quick Sort takes **1.13 seconds**, while Merge Sort takes **2.08 seconds**. As the input size increases, the difference becomes more noticeable. At **50,000 elements**, Quick Sort runs in **6.33 seconds**, whereas Merge Sort takes **11.36 seconds**. For **100,000 elements**, Quick Sort completes in **18.82 seconds**, while Merge Sort takes **23.10 seconds**.  

Both sorting algorithms have a time complexity of **O(n log n)**, but Quick Sort is usually faster because it sorts **in place**, meaning it does not need extra memory like Merge Sort. However, Merge Sort is more consistent since its performance does not depend on the choice of pivot like Quick Sort does.  

Overall, Quick Sort is the **better choice** for most cases due to its speed and efficiency, but Merge Sort is useful when **stable sorting** is needed.
