```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(10000, 100001, 10000)

# Read data from files
selection = np.loadtxt('SelectionSortTimes.txt')  
bubble = np.loadtxt('BubbleSortTimes.txt')       
insertion = np.loadtxt('InsertionSortTimes.txt') 
merge = np.loadtxt('MergeSortTimes.txt')         
quick = np.loadtxt('QuickSortTimes.txt')

plt.figure(figsize=(10, 7))

plt.plot(x, bubble, marker='o', c='r', label='Bubble Sort O(n²)')
plt.plot(x, insertion, marker='o', c='b', label='Insertion Sort O(n²)')
plt.plot(x, selection, marker='o', c='orange', label='Selection Sort O(n²)')
plt.plot(x, merge, marker='o', c='g', label='Merge Sort O(nlogn)')
plt.plot(x, quick, marker='o', c='y', label='Quick Sort O(nlogn)')

plt.title("Sorting Algorithm Running Times vs Input Size")
plt.xlabel("Input Size (n)")
plt.ylabel("Running Time (Seconds)")

plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```
