# Computation of Asymptotic Approximation

The primary task in this laboratory involves calculating the digit factorial sum of any given positive real number. This task is accomplished through the use of two main functions.

The first function, named "factorial", is responsible for calculating the factorial of a given number. In the context of this laboratory, the function specifically handles factorials for numbers ranging from one to nine (1â€“9).


*Factorial Function:*

```cpp
int factorial(int n)
{
    if(n == 0 || n == 1) return 1;

    int fact = n;
    for(int i = n - 1; i >= 2; i--)
        fact *= i;
    return fact;
}

The first calculation is the if statement, which acts as an exit condition and represents the best case O(1), contributing 2 primitive operations. The initialization of the fact integer contributes 1 operation.

Next, in the loop:
- The initialization (i = n - 1) contributes 1 operation.
- The comparison (i >= 2) runs 2(n - 2) times.
- The decrement (i--) executes 2(n - 1) times.

Inside the loop body:
- The fact operation contributes 2(n - 1) tasks.

Finally, the return statement runs 1 operation.

*Digital Factorial Sum Function:*

```cpp
std::vector<double> digit_fact_sum()
{
    long long int min_size = 1, max_size = 100000000000000;
    std::vector<double> result;

    for(long long int i = 1; i <= max_size; i *= 10)
    {
        std::cout << "Current Process: " << i << std::endl;
        auto start = std::chrono::high_resolution_clock::now();
        
        long long int sum = 0;
        for(long long int temp = i; temp > 0; temp /= 10)
            sum += factorial(temp % 10);

        auto end = std::chrono::high_resolution_clock::now();
        std::chrono::duration<double, std::milli> time_taken = end - start;
        result.push_back(time_taken.count());
    }
    return result;
}

The first step in the function, excluding the time logging functions (which are used to record the execution time), is the initialization of the minimum and maximum values for `n` (the number of inputs). In this case, the range spans from 1 to 100,000,000,000,000, with each subsequent value being the current number multiplied by 10. The main logic of the algorithm then proceeds with the inner `for` loop, which runs from `i` until `temp` becomes less than or equal to 0. The asymptotic complexity of this process is logarithmic with a base of 10.


### Following the given function here are is the solution for the theoritical asymptotic equation:
$$
1 + 1 + \log_{10}(n) + 1 + 2 \cdot \log_{10}(n) + 1 + 2 + 1(6n - 3 \cdot \log_{10}(n))
$$

$$
3 + \log_{10}(n) + 2 \cdot \log_{10}(n) + 2 + 3 + 6 \cdot n \cdot \log_{10}(n)

$$

### Best Case:
The best case is given by:

$$
\text{Best\_Case} = (\log_{10}(x) + 1) \cdot (\log_{10}(x) + 2)
$$

This suggests that the function has a logarithmic growth, with a slower rate due to the constants. As \(x\) increases, the complexity grows logarithmically, but at a reduced pace.

### Worst Case:
The worst case is given by:

$$
\text{Worst\_Case} = 5 + 3 \cdot \log_{10}(x) + 6 \cdot x \cdot \log_{10}(x)
$$

In the worst case, the complexity grows significantly faster due to the x * log base 10(n) term, indicating a combination of linear and logarithmic growth that increases quickly as x becomes large.
