This laboratory report examines three types of trees: **Binary Search Tree (BST)**, **AVL Tree**, and **Red-Black Tree**. Their performance is evaluated based on the time taken to execute key operations, including insertion, searching, and deletion. The analysis is conducted using input sizes ranging from **10,000** to **100,000**, increasing in increments of 10,000.

# Insertion Operation

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(10000, 100001, 10000)

# Read data from files
bst = np.loadtxt('BSTInsertionTimes.txt')  
avl = np.loadtxt('AVLInsertionTimes.txt')       
rbt = np.loadtxt('RBTInsertionTimes.txt') 


plt.figure(figsize = (10, 7))

plt.plot(x, bst, marker='o', c='r', label='Binary Search Tree (Non-Balancing)')
plt.plot(x, avl, marker='o', c='b', label='AVL Tree (Balancing)')
plt.plot(x, rbt, marker='o', c='g', label='Red-Black Tree (Balancing)')

plt.title("Balancing vs. Non-Balancing Trees")
plt.xlabel("Input Size (n)")
plt.ylabel("Running Time (Seconds)")

plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```
