This laboratory examines the performance of three key Minimum Spanning Tree (MST) algorithms - Kruskal's, Prim's, and Boruvka's - when applied to graphs of varying sizes.  The experiment tests these algorithms against randomly generated graphs with vertex counts ranging from 1000 to 2000, increasing by increments of 100.

# Minimum Spanning Tree Graph

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tabulate import tabulate

a = []
b = []
c = []
d = []

for line in open("MST_Time.txt", "r"):
    lines = [i for i in line.split(',')]
    a.append(float(lines[0]))
    b.append(float(lines[1]))
    c.append(float(lines[2]))
    d.append(float(lines[3]))

plt.figure(figsize = (10, 7))

plt.plot(a, c, marker = 'o', c = 'g', label = "Prim's")
plt.plot(a, b, marker = 'o', c = 'b', label = "Kruskal's")
plt.plot(a, d, marker = 'o', c = 'r', label = "Boruvka's")

plt.title("Kruskal’s, Prim’s and Boruvka’s Algorithm for Minimum Spanning Trees (MST)")
plt.xlabel("Size of Input")
plt.ylabel("Time (t) in seconds")

plt.legend()
plt.show()
plt.tight_layout()
```

## MST Algorithm Performance Analysis

### Kruskal's Algorithm
From the tests, Kruskal's algorithm was the slowest of the three MST algorithms. It took 1.587 seconds to process 1000 vertices and jumped up to 10.273 seconds for 1900 vertices. The results showed that as the graph got bigger, the time increased even faster. This makes sense because Kruskal's has to sort all the edges first, which takes longer when there are more edges to sort.

### Prim's Algorithm
Prim's algorithm was faster than Kruskal's but still much slower than Boruvka's in all the tests. The run times went from 1.174 seconds for 1000 vertices to 7.973 seconds for 1900 vertices. It was about 25-30% faster than Kruskal's algorithm, which is helpful but not amazing. Prim's works differently by growing a single tree from a starting vertex, which seems to give it a slight edge over Kruskal's.

### Boruvka's Algorithm
Boruvka's algorithm was tha fastest compared to the others. It only took 0.078 seconds for 1000 vertices and 0.353 seconds for 1900 vertices. That's about 29 times faster than Kruskal's and 22 times faster than Prim's at the largest size tested. The advantage of Boruvka's is that it processes multiple vertices at once, which seems to make it work much better for bigger graphs. Based on the results, Boruvka's is definitely the best choice if you need speed.

## Conclusion

*The results showed that Boruvka's algorithm was way faster than both Kruskal's and Prim's in all the tests. Prim's did better than Kruskal's, but Boruvka's blew them both away - it was like 29 times faster than Kruskal's when using the bigger graphs. The tests revealed that as the graph size increased from 1000 to 1900 vertices, the difference in speed got even bigger, which shows Boruvka's handles large graphs much better. So if someone is working with big graphs and needs something fast, they should probably go with Boruvka's algorithm based on what the results showed. Even though Boruvka's was fastest, Prim's and Kruskal's algorithms might still be useful for smaller graphs or when someone needs to visualize the step-by-step process of building a minimum spanning tree for learning purposes.*
